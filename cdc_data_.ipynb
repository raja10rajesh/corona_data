```
import pandas as pd
```

```
locations = pd.read_csv('locations.csv',low_memory=False)
total_cases = pd.read_csv('total_cases.csv',low_memory =False)
total_deaths = pd.read_csv('total_deaths.csv',low_memory=False)
```

```
print ("locations",len(locations))
print ("total_cases",len(total_cases))
print ("total_deaths",len(total_deaths))
```

```
total_cases.head()
```

```
final = total_cases.merge(total_deaths,on= 'date',how='left')
```

```
len(final)
```

```
popu_cases_death = locations.merge(final,how='left',left_on='location',right_on='date')
```

```
popu_cases_death.head()
```

```
popu_cases_death.to_csv('popu_cases_death.csv')
```

```
demo = pd.read_csv('domographic_world.csv',low_memory = False)
```

```
demo.head()
```

```
final = popu_cases_death.merge(demo,how='left',left_on='location',right_on='name')
```

```
final.head()
```

```
final_s = final.drop(['date','Unnamed: 0','name','pop2020_y','pop2020_y.1','pop2020_y.2','pop2020_x.2','pop2020_x.1','pop2020'], axis= 1)
```

```
final_s.to_csv('corona_final.csv')
```
